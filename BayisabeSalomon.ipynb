**Short Description**

The FAOSTAT Temperature Change dataset provides statistics on mean surface temperature change by country from 1961–2023. It reports monthly, seasonal, and annual anomalies relative to a 1951–1980 baseline climatology, along with standard deviations. Data are derived from NASA-GISS’s Global Surface Temperature Change (GISTEMP) records. Coverage includes 190 countries and 37 territories, with measurements expressed in °C.
