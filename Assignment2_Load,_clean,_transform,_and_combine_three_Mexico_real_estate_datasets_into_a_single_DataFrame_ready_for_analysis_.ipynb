***Assignment: Load, clean, transform, and combine three Mexico real-estate
datasets into a single DataFrame ready for analysis.***


```python
import pandas as pd
```

#Step 1: Read the three CSV files

```python
df1 = pd.read_csv("/content/mexico-real-estate-1.csv")
```

```python
df2 = pd.read_csv("/content/mexico-real-estate-2.csv")
```

```python
df3 = pd.read_csv("/content/mexico-real-estate-3.csv")
```

#Step 2: Inspect df1


```python
df1.shape
```

```python
df1.info()
```

```python
df1.head()
```

#Step 3: Clean df1

# Drop rows with any NaN value

```python
df1.dropna(inplace=True)
```

```python
df1.shape
```

# Remove "$" and "," from price_usd, then cast to float

```python
df1["price_usd"] = (df1["price_usd"]
                  .str.replace("$", "", regex=False)
                  .str.replace(",", "", regex=False)
                  .astype(float)
)
```

```python
df1["price_usd"].head()
```

#  Step 4: Clean df2
   

# Drop rows with any NaN value

```python
df2.dropna(inplace=True)
```

# Convert MXN → USD (exchange rate: 19 pesos per dollar in 2014)

```python
df2["price_usd"] = df2["price_mxn"]/19
```

```python
df2["price_usd"].head()
```

```python
df2.head()
```

# Drop the original MXN column

```python
df2.drop(columns=["price_mxn"], inplace=True)
```

```python
df2.head()
```

#  Step 5: Clean df3 – split "lat-lon" into two columns

# Drop rows with any NaN value

```python
df3.dropna(inplace=True)
```

```python
df3.head()
```

```python
df3.info()
```

# Split the "lat-lon" string on the comma separator

```python
df3[["lat","lon"]] = df3 ["lat-lon"].str.split(",", expand=True)
```

```python
df3.head()
```

```python
df3.info()
```

```python
df3["lon"] = df3["lon"].astype(float)
df3["lat"] = df3["lat"].astype(float)
```

#Step 6: Extract state name, drop unneeded columns

# State name is the 3rd segment (index 2) after splitting on "|"

# Format example: "México|Jalisco|Guadalajara|..."

```python
df3["state"] = df3["place_with_parent_names"].str.split("|").str[2]
```

```python
df3.head()
```

# Remove the original columns that are no longer needed

```python
df3.drop(columns=["place_with_parent_names", "lat-lon"], inplace=True)
```

#Step 7: Concatenate all three DataFrames

```python
df = pd.concat([df1,df2,df3], ignore_index=True)
```
