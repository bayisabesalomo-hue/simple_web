Research Question 1

**Which state has the most expensive real estate market?**

What are average mean price in each state

```python
mean_price=df.groupby("state")["price_usd"].mean().sort_values(ascending=False)
mean_price
```

```python
mean_price.plot(
    kind="bar",
    xlabel="state",
    ylabel="price usd",
    title="mean house price by state"
)
```

Price per size of meter

```python
df["price_per_m2"]=df["price_usd"]/ df["area_m2"]
df.head()
```

```python
mean_price_per_m=df.groupby("state")["price_per_m2"].mean().sort_values(ascending=False)
mean_price_per_m
```

```python
mean_price_per_m.plot(
    kind="bar",
    xlabel="state",
    ylabel="mean price m",
    title="Mean house price per m2"
);
```

# **Research** **Question** **2**

**Is there a relationship between home size and price?**

From our previous question, we know that the location of a home affects its price (especially if it's in Mexico City), but what about home size? Does the size of a house influence price?

```python
plt.scatter(x=df['area_m2'],y=df['price_usd'])
plt.xlabel("area_m2")
plt.ylabel("Price usd")
plt.title("Area vs price")
plt.grid(True);
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["area_m2"], df["price_usd"])
plt.xlabel("Area (m²)")
plt.ylabel("Price (USD)")
plt.title("Price vs Size")
plt.grid(True)
plt.show()
```
